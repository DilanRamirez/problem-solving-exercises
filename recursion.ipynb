```python


class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

class Solution:
    def Insert(self, T, newItem):
        if T == None:
            T =  TreeNode(newItem)
        elif T.val > newItem:
            T.left = self.Insert(T.left,newItem)
        else:
            T.right = self.Insert(T.right,newItem)
        return T

    def InOrderD(self, T, space):
    # Prints items and structure of BST
        if T is not None:
            self.InOrderD(T.right, space + '   ')
            print(space, T.val)
            self.InOrderD(T.left, space + '   ')
    
    #Problem 1: Range Sum of BST
    # Given the root node of a binary search tree, 
    # return the sum of values of all nodes with value between L and R (inclusive).
    # The binary search tree is guaranteed to have unique values.
    def rangeSumBST(self, root: TreeNode, L:int, R:int) -> int:
        if root is None:
            return 0
        return self.findNodes(root, L, R)

    # helper function
    def findNodes(self, root, L, R):
        rangeSum = 0
        if root is not None:
            # check if the node's value is inside of the range
            if L <= root.val and R >= root.val:
                # sum the value inside of the range
                rangeSum += root.val 
            # do the same process for the right and left side of the tree
            rangeSum += self.rangeSumBST(root.left, L, R) 
            rangeSum += self.rangeSumBST(root.right, L, R) 
        return rangeSum


    # problem 2: Invert Binary Tree 
    def invertTree(self, root: TreeNode) -> TreeNode:
        # check if the tree is not empty
        if root is not None:
            # assign left values to the right side, and rigth to the left side
            root.left, root.right = self.invertTree(root.right), self.invertTree(root.left)
            return root



    # problem 3: A binary tree is univalued if every node in the tree has the same value.
    def univaluedTree(self, root: TreeNode) -> bool:
        # if a tree is empty, it is univalued and it also means the left and right side had the same values.
        if not root:
            return True
        
        # check if the value of the root is the same of the right side, if not return False
        if root.right:
            if root.val != root.right.val: 
                return False
        # check if the value of the root is the same of the left side, if not return False
        if root.left:
            if root.val != root.left.val: 
                return False
        # recursive call for both side of the tree 
        return self.isUnivalTree(root.right) and self.isUnivalTree(root.left)
        

    # problem 4: Same Tree: Given the roots of two binary trees,
    # write a function to check if they are the same or not.
    # Two binary trees are considered the same 
    # if they are structurally identical and the nodes have the same value.
    def identicalTrees(self, root1: TreeNode, root2: TreeNode) -> bool:
        # if both trees are empty, it means both are identical. Also it means that the leaves are empty and also identical.
        if not root1 and not root2:
            return True
        # if the left or right leaf of the tree is missing, it means the trees aren't identical
        elif not root1 or not root2:
            return False
        # if the value of the roots are not the same, they are not identical
        elif root1.val != root2.val:
                return False
        # recursive call for both sides of the both trees 
        return self.identicalTrees(root1.right, root2.right) and self.identicalTrees(root1.left, root2.left) 

    # problem 5: Average of Levels in Binary Tree
    # Given a non-empty binary tree, return the average value of the nodes on each level in the form of an array. 
    def averageLevel(self, root: TreeNode):
        # In case the tree is empty 
        if root is None:
            return
        # Create the queue for BFS and the list of averages
        queue, averages, nodes_count_per_level,sum_per_level = [], [], 0, 0

        # Enqueue Root and initialize nodes_count_per_level and sum_per_level
        queue.append(root)
        # Append None as the root is alone
        queue.append(None)

        # BFS
        while(len(queue) > 0):
            node = queue.pop(0)
            if node is None:
                averages.append(sum_per_level / nodes_count_per_level)
                sum_per_level = 0
                nodes_count_per_level = 0
                queue.append(None)

                if queue[0] is None:
                    break
                else:
                    continue

            sum_per_level += node.val
            nodes_count_per_level += 1

            # Enqueue left child
            if node.left is not None:
                queue.append(node.left)

            # Enqueue right child
            if node.right is not None:
                queue.append(node.right)
        return averages

        ''' 
        2.	What prevented you from solving the prsoblem? 
            - I did not know how BFS works and had to implement in this problem.
        3.	What did you learn? Did you have to Google a little more to understand the solution? 
            - I learned how BFS works in BST.
        4.	What would you do differently in the future if you were presented with a similar problem?
            - I would see if BFS fits for the similar problem. If so, I would implement it.
        '''

    # A full binary tree is a binary tree where each node has exactly 0 or 2 children.
    # Write a function/method that, given the root of a binary tree, determines if it is full or not
    def fullBinaryTree(self, root):
        # if a node is empty or None, it is full
        if root is None:
            return True
        # if there are not leaves for a node or root, it is full
        if root.left is None and root.right is None:
            return True
        # if there are leaves, it checks the left and right side of the tree to see if the nodes are full or not.
        if root.left is not None and root.right is not None:
            return self.fullBinaryTree(root.left) and self.fullBinaryTree(root.right)
        return False

            


solution = Solution()
T,Y = None, None
A =[6, 3, 16, 11, 7, 17, 14]
B =[6, 3, 16, 11, 7, 17, 14]
for a in A:
    T = solution.Insert(T,a)
for b in B:
    Y = solution.Insert(Y, b)
solution.InOrderD(T, ' ')
print('******************')

# solution.rangeSumBST(root=T, L=3, R=14)
# inverted = solution.invertTree(root=T)
# solution.InOrderD(inverted, ' ')
# solution.univaluedTree(root=T)
# solution.identicalTrees(root1=T, root2=Y)
solution.averageLevel(root=T)
# solution.fullBinaryTree(T)
```

```python
# Problem7: splitArray that receives an array of ints as input
    # determines whether it is possible to divide the ints into two groups
    # the sums of the two groups are the same
def splitArray(inputList):
    # variables to store the addition values for each group and the index for traversing the list
    index, sumGroup1, sumGroup2 = 0, 0, 0
    # use the helper function to solve the problem
    return helperFunction(inputList, index, sumGroup1, sumGroup2)

# helper function
def helperFunction(inputList, index, sumGroup1, sumGroup2):
    # once all the list has been traversed, we compare the addition value of the two groups
    # and return if they are equal or not.
    if index >= len(inputList):
        return sumGroup1 == sumGroup2

    # use the current value of the list according to the index position to add it to the groupSum variable.
    currentValue = inputList[index]

    # Recursive call to find the sum of both group adding the current value to both groups.
    return helperFunction(inputList, index+1, sumGroup1+currentValue, sumGroup2) or helperFunction(inputList, index+1, sumGroup1, sumGroup2+currentValue)
    

inputList = [2,10,2,10]
splitArray(inputList)
```
