```python


class Solution:
    """844. Backspace String Compare"""
    def backspaceCompare(self, S: str, T: str) -> bool:
        S_stack, T_stack = [], []
        for char in S:
            if char == "#":
                if S_stack:
                    S_stack.pop(-1)
            else:
                S_stack.append(char)
        for char in T:
            if char == "#":
                if T_stack:
                    T_stack.pop(-1)
            else:
                T_stack.append(char)
                
        if T_stack == S_stack:
            return True
        return False

    def search(self, nums: List[int], target: int) -> int:
        '''704. Binary Search'''
        start = 0
        end = len(nums) - 1
        
        while start <= end:
            middle = (start + end) // 2

            if target == nums[middle]:
                return middle

            elif target > nums[middle]:
                start = middle + 1

            elif target < nums[middle]:
                end = middle - 1

        return -1

    def islandPerimeter(self, grid: List[List[int]]) -> int:
        '''463. Island Perimeter'''
        visit = set()
        def dfs(i, j):
            if i >= len(grid) or j >= len(grid[0]) or i <0 or j< 0 or grid[i][j] == 0:
                return 1

            if (i, j) in visit:
                return 0
            
            visit.add((i,j))
            perim = dfs(i, j+1)
            perim += dfs(i+1, j)
            perim += dfs(i, j-1)
            perim += dfs(i-1, j)

            return perim
        
        for i in range(len(grid)):
            for j in range(len(grid[0])):
                if grid[i][j] == 1:
                    return dfs(i, j)

    def isToeplitzMatrix(self, matrix: List[List[int]]) -> bool:
        '''766. Toeplitz Matrix'''
        for row in range(len(matrix)):
            for col in range(len(matrix[0])):
                if row > 0 and col > 0 and matrix[row][col] != matrix[row-1][col-1]:
                    return False
        return True

    def longestCommonPrefix(self, strs: List[str]) -> str:
        '''14. Longest Common Prefix'''
        if len(strs) > 200:
            return ""
        prefixDic = {}
        for word in strs:
            if len(word) > 200:
                return ""
            for letter in word:
                if letter not in prefixDic:
                    prefixDic[letter] = 1
                else:
                    prefixDic[letter] += 1
        prefix = ""
        for key in prefixDic:
            if prefixDic[key] == len(strs):
                prefix += key
        numberofOccurences = 0
        for word in strs:
            if prefix in word:
                numberofOccurences += 1

        if numberofOccurences == len(strs):
            return prefix
        else:
            return ""
```
