```python


class Solution:
    """844. Backspace String Compare"""
    def backspaceCompare(self, S: str, T: str) -> bool:
        S_stack, T_stack = [], []
        for char in S:
            if char == "#":
                if S_stack:
                    S_stack.pop(-1)
            else:
                S_stack.append(char)
        for char in T:
            if char == "#":
                if T_stack:
                    T_stack.pop(-1)
            else:
                T_stack.append(char)
                
        if T_stack == S_stack:
            return True
        return False

    def search(self, nums: List[int], target: int) -> int:
        '''704. Binary Search'''
        start = 0
        end = len(nums) - 1
        
        while start <= end:
            middle = (start + end) // 2

            if target == nums[middle]:
                return middle

            elif target > nums[middle]:
                start = middle + 1

            elif target < nums[middle]:
                end = middle - 1

        return -1

    def islandPerimeter(self, grid: List[List[int]]) -> int:
        '''463. Island Perimeter'''
        visit = set()
        def dfs(i, j):
            if i >= len(grid) or j >= len(grid[0]) or i <0 or j< 0 or grid[i][j] == 0:
                return 1

            if (i, j) in visit:
                return 0
            
            visit.add((i,j))
            perim = dfs(i, j+1)
            perim += dfs(i+1, j)
            perim += dfs(i, j-1)
            perim += dfs(i-1, j)

            return perim
        
        for i in range(len(grid)):
            for j in range(len(grid[0])):
                if grid[i][j] == 1:
                    return dfs(i, j)

    def isToeplitzMatrix(self, matrix: List[List[int]]) -> bool:
        '''766. Toeplitz Matrix'''
        for row in range(len(matrix)):
            for col in range(len(matrix[0])):
                if row > 0 and col > 0 and matrix[row][col] != matrix[row-1][col-1]:
                    return False
        return True

    def smallestPositive(self, A):
        # write your code in Python 3.6
        if len(A) < 1:
            return 1
        
        maxNun = max(A)
        minNum = min(A)
        memo = {}
        for num in range(minNum, maxNun+1):
            if num in A:
                memo[num] = 1
            else:
                memo[num] = 0
        for key in memo:
            if key < 0 and memo[key] == 0:
                return 1
            if memo[key] == 0:
                return key

        return maxNun + 1
    
    def BinaryGap(n):
        '''A binary gap within a positive integer N is any maximal sequence 
            of consecutive zeros that is surrounded by ones at both ends in the binary representation of N.

            For example, number 9 has binary representation 1001 and contains a binary gap of length 2.
            The number 529 has binary representation 1000010001 and contains two binary gaps: one of length 4 and
            one of length 3. The number 20 has binary representation 10100 and contains one binary gap of length 1.
            The number 15 has binary representation 1111 and has no binary gaps. The number 32 has
            binary representation 100000 and has no binary gaps
        '''
        one = False
        counter = 0
        longestGap = []

        binary = str(format(n, 'b'))
        for char in binary:
            if char == "1":
                one = False
                longestGap.append(counter)
                counter = 0
            elif char == '0':
                counter += 1
                one = True

        return max(longestGap)

    def CyclicRotation(n, k):
        '''
        Rotation of the array means that each element is shifted right by one index,
        and the last element of the array is moved to the first place
        '''
        counter = 0
        new = [0]*len(n)
        for i in range(len(n)):
            rotate = (i+k)%len(n)
            print(rotate)
            new[rotate] = n[i]
        return new

    def OddOccurrencesInArray(A):
        '''Find value that occurs in odd number of elements.'''
        memoPairs = {}
        for num in A:
            if num not in memoPairs:
                memoPairs[num] = 1 
            else:
                memoPairs[num] += 1
        
        for key in memoPairs:
            if memoPairs[key] % 2 != 0:
                return key
    
```
